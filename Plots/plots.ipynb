```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
def speed(model1,model2,name,name2):

 

    plt.figure(figsize=(10, 6))

    plt.plot(model1['Step'], model1['Value'], label=name, color='blue')

    plt.plot(model2['Step'], model2['Value'], label=name2, color='red')

    plt.title("Average Speed Comparison")
    plt.xlabel("Steps")
    plt.ylabel("Average Speed")
    plt.legend(loc="upper left")
    plt.grid(True)

    plt.show()
```

```python

def col(model1,model2,name,name2):
    
    plt.figure(figsize=(10, 6))

    plt.plot(model1['Step'], model1['Value'], label=name, color='blue')

    plt.plot(model2['Step'], model2['Value'], label=name2, color='red')

    plt.title("Collision Rate Comparison")
    plt.xlabel("Steps")
    plt.ylabel("Average Speed")
    plt.legend(loc="upper left")
    plt.grid(True)

    plt.show()
```

# Mappo vs MA2C

```python
model1 = pd.read_csv('mappo_speed.csv')  
model2 = pd.read_csv('ma2c_speed.csv')  
speed(model1,model2,"Mappo","MA2C")
```

```python
model1 = pd.read_csv('mappo_col.csv')  
model2 = pd.read_csv('ma2c_col.csv')  
speed(model1,model2,"Mappo","MA2C")
```

# MA2C vs Col100 e Col1000

```python
def speed3(model1,model2,model3,name, name2,name3):
    merged_data = model1.merge(model2, on='Step', suffixes=('_model1', '_model2')).merge(
        model3, on='Step', suffixes=('', '_model3')
    )
    merged_data.rename(columns={'Value': 'Value_model3'}, inplace=True)

    plt.figure(figsize=(12, 6))
    merged_data['Smoothed_model1'] = merged_data['Value_model1'].rolling(window=6).mean()
    merged_data['Smoothed_model2'] = merged_data['Value_model2'].rolling(window=6).mean()
    merged_data['Smoothed_model3'] = merged_data['Value_model3'].rolling(window=6).mean()

    plt.plot(merged_data['Step'], merged_data['Smoothed_model1'], label=name, color='blue')
    plt.plot(merged_data['Step'], merged_data['Smoothed_model2'], label=name2, color='red')
    plt.plot(merged_data['Step'], merged_data['Smoothed_model3'], label=name3, color='green')

    plt.title("Average Speed Comparison")
    plt.xlabel("Steps")
    plt.ylabel("Value")
    plt.legend(loc="upper left")
    plt.grid(True)

    plt.show()
    
    
    
    #plt.plot(merged_data['Step'], merged_data['Value_model1'], label=name, color='blue')

    #plt.plot(merged_data['Step'], merged_data['Value_model2'], label=name2, color='red')

    #plt.plot(merged_data['Step'], merged_data['Value_model3'], label=name3, color='green')
```

```python
def col3(model1,model2,model3,name, name2,name3):
    
    merged_data = model1.merge(model2, on='Step', suffixes=('_model1', '_model2')).merge(
        model3, on='Step', suffixes=('', '_model3')
    )
    merged_data.rename(columns={'Value': 'Value_model3'}, inplace=True)

    plt.figure(figsize=(12, 6))
    
    
    merged_data['Smoothed_model1'] = merged_data['Value_model1'].rolling(window=4).mean()
    merged_data['Smoothed_model2'] = merged_data['Value_model2'].rolling(window=4).mean()
    merged_data['Smoothed_model3'] = merged_data['Value_model3'].rolling(window=4).mean()

    plt.plot(merged_data['Step'], merged_data['Smoothed_model1'], label=name, color='blue')
    plt.plot(merged_data['Step'], merged_data['Smoothed_model2'], label=name2, color='red')
    plt.plot(merged_data['Step'], merged_data['Smoothed_model3'], label=name3, color='green')
    

    plt.title("Collision Rate Comparison")
    plt.xlabel("Steps")
    plt.ylabel("Value")
    plt.legend(loc="upper left")
    plt.grid(True)

    plt.show()
```

```python
model1 = pd.read_csv('ma2c_speed.csv')  
model2 = pd.read_csv('col100_speed.csv')  
model3 = pd.read_csv('col1000_speed.csv') 
speed3(model1,model2,model3,"MA2C","MA2C_col100","MA2C_col1000")
```

```python
model1 = pd.read_csv('ma2c_col.csv')  
model2 = pd.read_csv('col100_col.csv')  
model3 = pd.read_csv('col1000_col.csv') 
col3(model1,model2,model3,"MA2C","MA2C_col100","MA2C_col1000")
```
